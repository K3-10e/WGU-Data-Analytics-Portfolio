<h3>B. Reflection Paper</h3>
<h4>1. Dashboard Alignment</h4>
<p> &nbsp; &nbsp; As described in the data dictionary, the primary business goal is to “reduce
customer churn” which the provided dashboard accomplishes in analyzing (Western
Govenor’s University, n.d.). On the metrics tab of the visualization, the two graphs at the top
attempt to correlate a customer’s quality of life with the rate of churn in a particular region.
The region with the highest quality of life (59.05) had the lowest churn (25.12% of the total
region) whereas the region with the lowest (46.12) has one of the highest churns (27.78%)
so there is a potential connection between the two variables that could provide valuable
insight to the executives and potential decision making.

 &nbsp; &nbsp; Aside from analyzing factors that contribute to churn, it is important to consider
ways to retain current customers to reduce churning. For example, the bottom left graphs
measure how customer and business engagement might impact the bandwidth a customer
might use as well as how equipment failure impacts communication between the parties.
Knowledge on why users leave and why users stay could assuredly “guide [executive
leaders] decision-making” as specified in the data dictionary. Furthermore, the
presentation of data is “broad and understandable” in that sections are appropriately
color-coded with clear labeling and provide greater detail when interacted with.
</p>
<h4>2. Additional Data Set Insights</h4>
<p> &nbsp; &nbsp; The additional data set, ‘qol_states_2024.csv’, provides numeric scores for each
state in various sections of life. This includes how safe one’s surroundings are, the
affordability of things, the quality of education and health care, and the overall state of the
economy in the area. I selected this additional data set because I want to know if any
exterior aspect of a customer’s life affects their business with the telecommunications
company. For example, some service add-ons could be seen as an unnecessary luxury to
customers who have tighter budgets. The bottom right graph under the metrics tab
highlights customers quality of life in affordability and how it relates to different services a
customer uses. As a result, those that do not have these extra services on average have a
slightly lower score in affordability. Knowing the background of a customer, such as their
financial and overall wellbeing, will help the business adjust their strategies to better
support their customers.</p>
<h4>3. Decision-Making Support</h4>
<p> &nbsp; &nbsp; The regional map tab of the dashboard features a heatmap of regions across the
United States that indicate the concentration of customers. Hovering over different regions
also provides a count of customers where the New England region has the least customers
at 543 and the South Atlantic region has the most at 1642. With this information, business
leaders can utilize different marketing strategies by focusing on customer acquisition in
regions with less customers and concentrating on activities that promote loyalty for areas
with many customers.

 &nbsp; &nbsp; As mentioned in the previous section, the chart on the effects of affordability on a
customer’s services is an important analysis that can be used for decision making. While it
can show which services might not be as popular on a tighter customer budget, it can also
explain which services customers want regardless of their financial situation. These are the
essential services that the business can market or bundle together to attract more
customers.
</p>
<h4>4. Interactive Controls</h4>
<p> &nbsp; &nbsp; On the regional map, users can click on one of the regions to filter the below graphs
for data just from that region. Clicking that highlighted region again will display the data
overall once again. This feature is included with idea that if executives wanted to market to
specific regions, like mentioned in the previous section, they could have a greater
understanding of the demographic of these specific customers. This functionality is also
useful to the regional vice presidents since the data grouping aligns with their operational
responsibilities.

 &nbsp; &nbsp; The ‘Visible Chart’ options on the right side of the screen create the biggest
difference in the presentation of the data. The charts on the regional map are mainly
descriptive whereas charts on the metrics tab are more analytical. Overall, executives wish
to one, understand who their customers are, and two, understand what factors lead to
churn or retention. The data is sectioned according to these objectives and will allow the
leaders to switch between goals by clicking the appropriate bubble option.
</p>
<h4>5. Colorblindness</h4>
<p> &nbsp; &nbsp; Tableau features a colorblind color palette when coloring your graph which is useful
for individuals with this disability. Ivan Kilin in “The best charts for color blind viewer”
recommends “[using] direct labels instead of a legend” for graphs that feature multiple
categories. In the regional map tab of the dashboard, I combine these principles by labeling
slices of the pie charts, as well as using the colorblind friendly palette. Additionally, I use
one hue color for the customer heatmap which makes “the chart readable for all kinds of
color blindness” (Kilin, 2022). And for just one more layer of security, hovering over any
element of the charts highlights the element and provides a tooltip that describes what the
user is looking at. So even if the viewer has a hard time distinguishing colors, they can
understand the graph by hovering over its elements.</p>
<h4>6. Data Representations</h4>
<p> &nbsp; &nbsp; On the bottom left corner of the metrics tab, there are two charts that compare
business and customer engagement to a customer’s equipment failure and their use of
bandwidth. Knowing that the company wants to increase retention, I thought that analyzing
the effectiveness of communication would be a good starting point. I selected a user’s
bandwidth because all customers use this service to some extent so it would be a good
measure seeing if engagement influences usage. In contrast, I wanted to see if
communication itself was affected by a customer’s yearly equipment failure as I figured
failing devices would be a potential reason for an increase in communication. I compared
these two factors to the customer contacting the business or the business sending emails
to the customer depending on which dropdown value is selected. While communication
does not influence a user’s bandwidth, there is an increase in communication for
customers whose equipment has failed more than four times. These data representations
help provide a clue to the business leaders into what a customer wants from the business
and which customers want more interactions.</p>
<h4>7. Audience Analysis</h4>
<p> &nbsp; &nbsp; I created both tabs of the dashboard with all executives in mind. Both the Senior
Vice President for Customer Experience (SVP) and the Executive Vice President of Sales
(EVP) are interested in the demographics of customers, with the EVP and the Panel of
Regional Vice Presidents (Regional VP) showing particular interest in data by region. To
account for all of these needs, the regional map tab presents demographical data and
allows the viewer to filter information by any particular region they may be curious about.
Another point of focus for the SVP and the EVP is improving recruitment and retention. I
make sure to cover potential points of insight regarding these factors in the metrics panel
of the dashboard. As described in previous sections, I focused on the idea of engagement
and its relation to business activities as well as considered factors that could hinder
retention such as one’s quality of life. The dashboard is suited to meet every business
leader’s needs, so it is a valuable tool for decision making.</p>
<h4>8. Universal Access</h4>
<p> &nbsp; &nbsp; The presentation was built to be as clear and simple as possible so it may be
accessible to all audiences. Uploading the dashboard onto Tableau Public enables
interested audiences to discover information without needing to install the application and
download the dashboard as detailed in part A; all the audience needs is a clickable link.
Stated in a previous section, all charts feature colorblind friendly colors as well as
descriptive tooltips when hovering over data to promote accessibility. Additionally, there
are labeled axis on every chart and titles on the metrics tab which promote ideas on how to
interpret the charts. No extra technical knowledge is needed to navigate the dashboard so
viewers that find themselves less than tech-savvy will still access it easily.</p>
<h4>9. Effective Storytelling</h4>
<p> &nbsp; &nbsp; The graph featuring the churn percentage by region as well as the quality of life
score by region are two graphs that are intended to be viewed together. When you click on a
bar in one graph, a bar on the other graph with the same region will be highlighted. The goal
being to relate a customer’s sense of well being to their choice of continuing business with
the telecommunications company. Analyzing these graphs do show some sense of
correlation as the high quality of life region had one of the least churns and the lowest
quality of life had the highest churn. This information can help the business predict which
customers are more volatile and which they can expect to retain for a longer period of time.

 &nbsp; &nbsp; Also on the metrics page, a customer’s quality of affordability contrasted against
what services they have is intended to show the true value of different services. If a service,
whether the customer has it or not, has similar scores in affordability, the service is valued
separate from the consumer’s budget and represents a more desirable feature. If a
customer having the service indicates a higher quality of life than not having the service,
then the service can be considered a luxury, not a critical aspect. This keys the audience in
to understanding which services are important and which are just bonus features for those
that can afford it.
</p>
