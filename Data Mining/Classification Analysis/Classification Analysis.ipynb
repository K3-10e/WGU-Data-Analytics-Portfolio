<h2>Part II: Method Justification</h2>
<h3>B. Justification and Reasoning</h3>
<h4>1. Explanation of Classification Method</h4>
<p>&nbsp; &nbsp; The k-nearest neighbors algorithm works by viewing characteristics of a target data
point (the independent variables), finding other data points with similar values, and then
calculates the mode value of the selected characteristic (the dependent variable) from
these other data points. This resulting value is the dependent value for the data point in
question. For this project, I will be using thirteen variables alongside the dependent
variable (all described in C2) in the KNN model. From the functions provided by various
Python libraries, I expect to discover the effectiveness of the model in accurately predicting
whether a customer uses online security services and the success rate of prediction in the
training and testing data itself.</p>
<h4>2. Summary of Method Assumption</h4>
<p>&nbsp; &nbsp; For the KNN model, it must be assumed “that similar points can be found near one
another” (elastic, n.d.). Since a data point’s k-neighbors are used to predict its value, it is
important that these neighbors contain similar kinds of data as it provides logistical
justification for why a certain value is being predicted. Using random neighbors for feature
prediction will severely affect the model’s effectiveness resulting in a model that is no
better than wildly guessing the value of a data point. Organizing values by likeness is the
core basis for the KNN method to function.</p>

<h4>3. Steps for Analysis</h4>
<p>&nbsp; &nbsp; To prepare the data for analysis, the records first must be cleaned and then
formatted (transformed) for the KNN algorithm. The cleaning process entails detecting and
treating duplicates, missing values, and outliers. Following that, formatting the data will
include removing columns irrelevant to the analysis, converting all values into numeric
values as describe in C1 and normalizing numeric values.

&nbsp; &nbsp; In terms of detecting entire row duplications, the length of the unaltered database
subtracted from the length of the database with duplicates removed indicates how many
row duplicates there are: zero. For columns themselves, only ‘CaseOrder’, ‘Customer_id’
and ‘Interaction’ require no duplicated values which can be detected with a sum of the
‘duplicated()’ function. Since this operation returns zero, there are no duplicates in the
database.
</p>

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, roc_curve, auc, roc_auc_score
from sklearn.preprocessing import StandardScaler

class text:
   BOLD = '\033[1m'
   UNDERLINE = '\033[4m'
   END = '\033[0m'

df = pd.read_csv('churn_clean.csv', keep_default_na = False, 
                 na_values = [' ', '#N/A', '#N/A N/A', '#NA', '-1.#IND', '-1.#QNAN', '-NaN', '-nan', '1.#IND', '1.#QNAN',
                             '<NA>', 'N/A', 'NA', 'NULL', 'NaN', 'n/a', 'nan', 'null'])
# aligns the df index to match the index of CaseOrder so df[0] doesn't exist
df.index = df.index + 1

print (text.BOLD + "\n-- Begin Data Analytics --\n" + text.END)
# region Prepare Data
# Duplicates
def detect_duplicates(col_name):
   """Detect Duplicates: 
      Detects duplicate values in a single column.
   """
   print(f"{col_name}: ",  df.duplicated(subset = col_name).sum())
print(text.UNDERLINE + "Detecting Duplicates" + text.END)
print("Exact row: ", len(df)-len(df.drop_duplicates()))
detect_duplicates('CaseOrder');
detect_duplicates('Customer_id');
detect_duplicates('Interaction');
df.drop_duplicates()
```

<p>&nbsp; &nbsp; Detecting outliers requires analyzing the values of the data and using one’s best
judgment. With this in mind, all numeric values of the set are sorted by the source of the
data, either from a census, sign-up information, or data from the business itself. Next, the
‘describe()’ function is applied to every numeric column which shows the mean, standard
deviation and the values at certain percentages of data (quintiles). The results of this code
showed all but one column contained reasonable values, with the problematic column
being ‘Income’. Having a minimum of 348.67 and a maximum of 258,900.70, these
amounts are a suspiciously large range. Furthermore, this variable comes from a sign-up
values so the chances for incorrect information being reported is increased compared to a
census value or business generated data. To treat this issue, the interquartile range method
with a threshold of 1.5 was applied and returned 336 rows of outlying values. Since this is
only three percent of all rows, these outliers are removed from the data set.</p>

```python
def describe_vars(list_name, list):
   """Describe Vars: 
      Used to describe variables and to visually check for outliers. 


      Compare min and max to each other.
      Compare if 25% is close to min and if 75% is close to max.
      Consider where the information was obtained.
   """
   print(f"{list_name} Values")
   print(str(df[list].describe()) + "\n")
def treat_outliers(col):
   """Treat Outliers: 
      Used to treat outliers using IQR. 
   """
   q1 = df[col].quantile(0.25)
   q3 = df[col].quantile(0.75)
   iqr = q3 - q1
   outliers = df[(df[col] < q1 - 1.5 * iqr) | (df[col] > q3 + 1.5 * iqr)]
   print(f"Outlier Count for {col}: {len(outliers)}")
   df.drop(outliers.index, inplace=True)
   print(f"New dataframe length: {len(df)}\n" )
   print(str(df[col].describe()))
# Outliers
census_list = ['Population']
signup_list = ['Age', 'Children', 'Income']
business_list = ['Outage_sec_perweek', 'Email', 'Contacts', 'Yearly_equip_failure', 
                'Tenure', 'MonthlyCharge', 'Bandwidth_GB_Year']
print(text.UNDERLINE + "\nDetecting Outliers" + text.END)
describe_vars("Census", census_list);
describe_vars("Sign-up", signup_list);
describe_vars("Business Generated", business_list);
print(text.UNDERLINE + "Treating Outliers" + text.END)
treat_outliers('Income')
```

<h2>Part IV: Analysis</h2>
<h3>D. Data Analysis and Results</h3>
<h4>1. Splitting the Data</h4>
<p>&nbsp; &nbsp; The data was split 80% into training (7731 rows) and 20% into testing (1933 rows).
These values can be found in the following csv files: ‘Xtrain.csv’, ‘Xtest.csv’, ‘ytrain.csv’, and
‘ytest.csv’ where ‘X’ represent the independent variables and ‘y’ is the dependent.</p>
<h4>2. Output and Intermediate Calculations</h4>
<p>&nbsp; &nbsp; To analyze the data set, the k-nearest neighbors (KNN) method was utilized. KNN
works by finding datapoints with similar independent variables to a certain data point and
determines the mode of the dependent variables from these data points. The mode is then
applied to the target data point as its dependent variable. To determine the optimal number
of neighbors to use in the model, the ‘GridSearchCV()’ function from the
sklearn.model_selection library was used. This tests attempts KNN with different numbers
of neighbors and calculates how many neighbors results in the most accurate model.
Testing a range of 1- 50 for this model indicated that k = 46 neighbors produced the greatest
accuracy. For all options in the KNN model, I used the default settings provided by the
sklearn.model_selection library. No intermediate calculations were performed for this
section.</p>


```python
# region KNN Analysis
print(text.UNDERLINE + "\nKNN Analysis" + text.END)
# split dataset [In-Text Citation: (GeeksforGeeks, 2023)]
X = prepared_df.drop(['OnlineSecurity_Yes'], axis=1)
y = prepared_df[['OnlineSecurity_Yes']]
X_train, X_test, y_train, y_test = train_test_split( 
             X, y, test_size=0.2, random_state=1001)
X_train.to_csv('Xtrain.csv', index=False)
X_test.to_csv('Xtest.csv', index= False)
y_train.to_csv('ytrain.csv', index= False)
y_test.to_csv('ytest.csv', index= False)

# transforms the y test and train values to an array for knn and roc calculations
y_test_val = y_test['OnlineSecurity_Yes'].values
y_train_val = y_train['OnlineSecurity_Yes'].values

# find best k-value [In-Text Citation: (Korstanje, n.d.)]
parameters = {"n_neighbors": range(1, 50)}
k_score = GridSearchCV(KNeighborsClassifier(), parameters)
k_score.fit(X_train, y_train_val)
print(f"{k_score.best_params_}\n")

# perform knn [In-Text Citation: (GeeksforGeeks, 2023)]
knn = KNeighborsClassifier(n_neighbors= 46)
knn.fit(X_train, y_train_val)

# confusion matrix
print("Confusion Matrix")
y_preds = knn.predict(X_test)
y_confusion = pd.crosstab(columns = y_test_val, colnames = ["Actual"], 
                          index = y_preds, rownames = ["Predicted"], margins=True, margins_name="Total")
print(y_confusion)

# classification report [In-Text Citation: (Bobbitt, 2022)]
print("\nClassification Report")
print(classification_report(y_test, y_preds))

# test/training accuracy
training = knn.score(X_train, y_train)
testing = knn.score(X_test, y_test)
print(f"Training Accuracy: {training}, ({round(training*100, 1)} %)")
print(f"Test Accuracy: {testing}, ({round(testing*100, 1)} %)")

# AUC score [In-Text Citation: (GeeksforGeeks, 2024)]
auc_score = roc_auc_score(y_test, y_preds)
print(f"Area Under Curve Score: {auc_score}, ({round(auc_score*100, 1)} %)")

# ROC curve/graph [In-Text Citation: (GeeksforGeeks, 2024a)]
y_pred_proba = knn.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test_val, y_pred_proba)
roc_auc = auc(fpr, tpr)
plt.figure()  
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr)
plt.title('ROC Curve for KNN')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.show()
```

<h2>Part V: Data Summary and Implications</h2>
<h3>E. Data Analysis Summary</h3>
<h4>1. Accuracy and AUC</h4>
<p>&nbsp; &nbsp; The training accuracy of the model was 0.642 meaning about 64.2% of all
predictions made on the training dataset were correct in determining whether a customer
uses online security. The testing accuracy was also a similar rate, with 64.0% of predictions
being accurate. Both values aren’t very high and call the model’s predictability into
question. In terms of the area under the curve(AUC) discovered from the receiver operator
characteristic (ROC), the value came out as 0.5055. This value indicates the KNN
classification model is only 50.6% accurate at predicting a customer’s online security
usage. Given that there are only two possible options for online security, either a client has
it or they do not, a prediction rate around 50% indicates the model is no better than just
guessing a customer’s status. Overall, the model is not suitable for making predictions.</p>
<h4>2. Results and Implications</h4>
<p>&nbsp; &nbsp; Connecting back to the question posed at the beginning of this project, the testing
results indicate that a customer’s characteristics cannot determine if they use the online
security add-on. As the AUC value is barely higher than 0.5, the model is only slightly better
than a “random assignment” of values (Dash, 2022). One potential path to model
improvement includes selecting more demographical variables such as ‘Children’, ‘Area’,
or ‘Marital’ which could provide a bigger picture of understanding for the model. Another
important area of improvement can come from better hyperparameter tuning. As I am new
to the concepts of machine learning, I allowed the KNN model to use default parameters
such as uniform weighted values, the minkowski metric for distance, and an auto-selected
algorithm for finding the nearest neighbors (scikit-learn, n.d.). With more experience I can
select better parameters to use in the future. For the selection of the number of neighbors
to use, I limited the ‘GridSearchCV()’ function to test up to 50 neighbors as the calculation
increases in time with higher values. Increasing the number of neighbors for this function
could also contribute to a more precise model.</p>
<h4>3. Limitation</h4>
<p>&nbsp; &nbsp; As indicated by the ‘churn’ variable, this data set only provides current customers or
customers that have just canceled their business with the company. This is fine for
predicting current trends or features about a client, however, it is harder to get a long-term
idea about customers in general. I think this data analysis could have entirely different
results if the records included clients that have canceled beyond the past month. Another
limitation comes from the distribution of data for the dependent variable. 6,424 records of
the 10,000 do not have online security. While this is close to an even distribution, there is
still clearly a skew towards not having this service. This makes it harder to predict
customers that do use this service and increases the chance of these customers being
classified as not using this service.</p>
